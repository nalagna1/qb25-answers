```python
reads = [
    'ATTCA', 'ATTGA', 'CATTG', 'CTTAT', 'GATTG', 'TATTT', 'TCATT',
    'TCTTA', 'TGATT', 'TTATT', 'TTCAT', 'TTCTT', 'TTGAT'
]

edges = set()
k = 3

for read in reads:
    for i in range(len(read) - k + 1):
        kmer = read[i:i+k]
        prefix = kmer[:-1]
        suffix = kmer[1:]
        edges.add(f"{prefix}->{suffix}")

# Write to file, one edge per line, format: ATT -> TTC
with open('edges.txt', 'w') as f:
    for edge in sorted(edges):
        f.write(f"{edge}\n")

with open('debruijn.dot', 'w') as f:
    f.write('digraph {\n')
    for edge in sorted(edges):
        src, dst = edge.split('->')
        f.write(f'  {src.strip()} -> {dst.strip()};\n')
    f.write('}\n')

# Bash commands: conda create -n graphviz -c conda-forge graphviz
#                conda activate graphviz
#                dot -Tpng debruijn.dot -o ex2_digraph.png
```
