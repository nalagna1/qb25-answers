```python
import numpy as np

GENOME_SIZE = 1_000_000
READ_LENGTH = 100
COVERAGE_X = 10
READS_NEEDED = GENOME_SIZE * COVERAGE_X // READ_LENGTH  # 100,000

coverage = np.zeros(GENOME_SIZE, dtype=int)
for _ in range(READS_NEEDED):
    start = np.random.randint(0, GENOME_SIZE - READ_LENGTH + 1)
    coverage[start : start + READ_LENGTH] += 1

np.savetxt("coverage_10x.txt", coverage, fmt='%d')
```
